```python
import pandas as pd
df_sales = pd.read_excel('08_linechart.xlsx')
df_sales
```

```python
from matplotlib import pyplot as plt

# for the figure size
plt.figure(figsize=(12,5))

plt.plot(df_sales['Quarter'], df_sales['Fridge'],color='red', label='Fridge') # it always need 2 axis the x and y axis
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'],color='green',label='Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'],color='blue',label='Washing Machine')

plt.title('Product Sales')
plt.ylabel('Revenue (mln $)')
plt.xlabel('Financial quarter')

plt.legend()

plt.show()
```

```python
total_sales = df_sales[['Fridge','Dishwasher','Washing Machine']].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', explode=(0,0.1,0),shadow=True, startangle=90)

plt.show()
```

```python
df_sales.plot(kind='bar', x='Quarter')
plt.xticks(rotation=45)

plt.show()
```

```python
df_score = pd.read_excel('08_histograms.xlsx')
df_score.head()
```

```python
plt.hist(df_score['Exam_Score'])
plt.show()
```

```python
# for histograms the seaborn librart is better
```

```python
import seaborn as sns
# 1st method
sns.histplot(df_score['Exam_Score'], kde=True)
# kde stands for kernal density

plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Exam Marks')
plt.show()
```

```python
# 2nd method

sns.histplot(data= df_score, x='Exam_Score', kde=True)
# kde stands for kernal density

plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Exam Marks')
plt.show()
```
