```python
import pandas as pd
```

```python
pakistan_weather = pd.DataFrame({
    'city':['Peshawar','Islamabad','Karachi'],
    'temperature':[19,17,25],
    'humidity':[64,80,55]
})
pakistan_weather
```

```python
uk_weather = pd.DataFrame({
    'city':['London','Birmingum','Lankashire'],
    'temperature':[3,11,19],
    'humidity':[85,69,57]
})
uk_weather
```

For merging we can use concate function

```python
df = pd.concat([pakistan_weather, uk_weather])
df
```

```python
# as we see the index are not specified correctly so for that
df = pd.concat([pakistan_weather, uk_weather], ignore_index=True)
df
```

```python
# we can also speicify keys like pakistan and Uk for these columns:
df = pd.concat([pakistan_weather, uk_weather], keys=['Pakistan','UK'])
df
```

```python
# the benefit of this key is that later on we can access it:
df.loc['UK']
```

```python
# we can also concatenate these 2 DF by index, Let us create a new DataFrame
temp_df = pd.DataFrame({
    'city':['Karachi','Lahore','Peshawar'],
    'temperature':[33,29,20]
},index=[0,1,2])
temp_df
```

```python
windspeed_df = pd.DataFrame({
    'city':['Peshawar','Lahore'],
    'windspeed':[6,9]
},index=[2,1])
windspeed_df
```

```python
# if we dont use axis then it will return same index value multiple time which will be an exception
pd.concat([temp_df, windspeed_df],axis=1)
```

SQL Join function and pandas merge function similarities and understanding
# Example

```python
df1 = pd.DataFrame({
    'city':['new york','chichago','orlando','baltimore'],
    'temperature':[21,14,35,38]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['chichago','new york','san diego'],
    'humidity':[65,68,71]
})
df2
```

```python
# usual merge
pd.merge(df1,df2, on='city', how= 'inner')

# in usual merge we only got same cities data, it is same as SQL inner join in which we only get data for same parameters
```

```python
# left join
pd.merge(df1,df2, on='city', how= 'left')

# in left join we will get all data from the left side and same data if present in right that we will get only from right side.
```

```python
# right join
pd.merge(df1,df2, on='city', how= 'right')

# similarly in right join we will get all the data from right side but get only the matching data from left
```

```python
# full outer join
pd.merge(df1,df2, on='city', how= 'outer')

# in full outer joiner we will get everything from both left and right side
```
