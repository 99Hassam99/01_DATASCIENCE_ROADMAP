The main difference between the mean and the median is how they are calculated and what they represent: 

**Mean:**
The **mean** is the average of a set of numbers, calculated by adding all the numbers together and dividing by the total number of values. The mean is also known as the arithmetic mean. 

**Median:**
The **median** is the middle value of a set of numbers, ordered from least to greatest. If there is an even number of values, the median is the average of the two middle numbers. 

The mean is a good representation of the typical value in a set of numbers.
The median is a good measure of central tendency for skewed distributions or distributions with outliers. For example, the median is often used to represent income distributions, which are usually highly skewed. 

```python
df = pd.read_csv('04_weather_data.csv')
df
```

```python
df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No Event'
})
```

```python
df.temperature.mean()
```

```python
df.windspeed.mean()
```

```python
df.temperature.median()
```

```python
df.windspeed.median()
```

```python
df.fillna({
    'temperature': df.temperature.median(),
    'windspeed': df.windspeed.median(),
    'event': 'No Event'
})
```

There is another method called **ffill** **Forward Fill**.
It will look at the previous value. Like it will forward the previous value

```python
df
```

```python
df.ffill()
```

As we see it word but if we look at index **5** and **6** it forward the value multiple times and it will forward infinite times if the cells are empty.
so we can do limitation.

```python
df.ffill(limit=1)
```

# See The index 5 and 6 now it only forward the value one time

Similar to **Forward fill** there is another method called **BFILL** means **Backward Fill**.
It will return the value from the upcoming value like the front value.

```python
df
```

```python
df.bfill()
```

By Default the value of **Bfill** is on rows like it fill **row** wise but we can change it to **columns**.

```python
df
```

```python
df.bfill(axis='columns')
```

Sometimes the NA value are less in number i-e 2 or 4 or 10 values, so instead of creating NA values we can simply drop them.

```python
df
```

```python
df.dropna()
```

```python
df.dropna(how='any') # means drop the row with one NA value 
```

```python
df.dropna(how='all') # means drop the row with all the columns to be NA value 
```
