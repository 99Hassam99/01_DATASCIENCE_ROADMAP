# What is feature engineering?

Feature engineering is a very process in the field of data analysis and machine learning.



There was a study publish by forbes magazines on what data scientist spend most of thier time on and from this chart it was very clear that they spend majority of thier time in data cleaning because:



Because in the real world the data looks pretty messy as shown in this picture here and we need to clean that data so that we can feed that clean data to machine learning models. ML models will not work if the data is messy, even if we are not using ML and doing simple data analysis then still we need clean data so data cleaning is very important and that is where data analyst and data scientist spend most of thier time on and feature engineering is something that helps us with data cleaning.

For example:



Lets assume we are studying the home prices and these are the home prices in the city of Banglore in India. To do the analysis on this one the things that we can do is that we can create a new column which will be price per sqrft. Now this dataset can be coming from internet or from a variety of sources and it can have errors, so by creating this new column we are trying to analyze if we have any data errors and once we have this new column now we can use either of our domain knowledge, so if we have a business manager that we are working closely with he would say that in Banglore we cannot get a home at a rate of price per sqrft equal to 3500 sqrft so when we see this particular datapoint highlighted which has price per sqrft as 500 it clearly looks like an error and this error we detected by using a domain knowledge, we can also use visualization and by plotting bar chart this datapoint 500 is cleary comes out to be a different than the rest of the data points, we can also use mathematics and statistics where we can use techniques like to standard deviation to detect these kind of spatial obsevation, now this data point here is called an **Outlier** and these 3 techniques listed  on the right hand side are called the **Outlier Detection Techniques**.



If we throw back to pandas we also done how to handle missing values, sometimes when we receive the incoming dataset it might have missing values, here for these 2 data points the bedrooms are missing so what we can do is to just take a median which will be 3 and fill those missing values so **handling missing** values is also very important during the process of data cleaning.



We also had cover one hot encoding in machine learning series where we replace the text data with numeric data, here in this image the text data is town and other are property prices in different towns in state of new jersey and when we want to build an ML model the model doesnt understand text so we have to convert that into numeric data and we converted that town data into last 3 columns consisting of 0s and 1s and this process is called **One Hot Encoding**



So these 3 techniques that we covered which is:

1. Outlier detection
2. One hot encoding
3. Handling missing values

These are the few examples of **Feature engineering**.
