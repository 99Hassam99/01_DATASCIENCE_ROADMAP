So we have covered this concept that whenever we have an overfit model it is likely that we get high variance and when we have underfit model it is likely that we will get high bias and when we have balance fit we will get low variance and low bias and as a data scientist we want to comeup with a model that had a balance fit.

Now there is a bulls eye diagram as shown:



Where the central circle represent the truth. This is the truth samples and when the predicted value like here the white small diamonds are the predicted values when they are near to the inner circle which is the truth then it is called low bias and bias is all about how close you are to the truth so if you are closer to the middle circle you have low bias.

If you are far away from the middle circle you have high bias but in both the cases on the left hand side  these diamonds they are clustered together such that they are like a close family that are together so whenever they are together it is called low variance and in high variance they are apart as shown in the right hand side figure but they are still near to the truth in one diagram which is the inner circle hence there is a low bias.

Worst case scenario is a high bias where we are far away from the truth which is this middle circle that means the bias is high and also these individual predicted values they are separated out hence we have a high variance.

How do we get a balance fit model so there are several techniques.



1. The first one is called cross validation
2. The 2nd one is L1 and L2 regularizarion
3. Dimensionality reduction (PCA)
4. Ensemble techniques (Bagging boosting)
