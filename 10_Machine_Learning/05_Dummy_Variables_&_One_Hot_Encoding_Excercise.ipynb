```python
import pandas as pd
df = pd.read_csv('05_carprices.csv')
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df['Car Model'],df['Sell Price($)'],color='blue',label='car prices')
plt.xlabel('Car model')
plt.ylabel('Price')
plt.title('Scatter Plot Car Prices')
plt.show()
```

```python
dummies = pd.get_dummies(df['Car Model'])
dum = dummies.astype(int)
dum
```

```python
merged = pd.concat([df,dum],axis='columns')
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
x = final.drop('Sell Price($)',axis='columns')
x
```

```python
y=final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model2 = LinearRegression()
```

```python
model.fit(x,y)
```

```python
#Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

model.score(x,y)
```

```python
#Predict price of a mercedez benz that is 4 yr old with mileage 45000

model.predict([[45000,4,0,0]])
```

```python
#Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```
