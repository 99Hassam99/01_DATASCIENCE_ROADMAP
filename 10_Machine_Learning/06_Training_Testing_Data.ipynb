So somwtimes we train the model using the entire dataset but that's not a good strategy, the good strategy is to split the set into parts where you use part of the dataset for actual training and use the remaining dataset for testing the model as shown:

The reason is that you want to use those samples for testing the models that model has not seen before so for example in above figure if we use the first 8 samples to train the model and then use the remaining 2 for testing the model then i will get a good idea on the accuracy of model because the model has not these 2 samples before.

```python
# load module
import pandas as pd
```

```python
# load dataset
df = pd.read_csv('06_carprices.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**Car Mileage vs Sell Price**

```python
plt.scatter(df['Mileage'],df['Sell Price($)'])
```

**Car Age vs Sell Price**

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
```

```python
X = df[['Mileage','Age(yrs)']]
Y = df[['Sell Price($)']]
```

```python
X
```

```python
Y
```

```python
# Train test split method

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.1) #,random_state=10
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.predict(X_test)
```
