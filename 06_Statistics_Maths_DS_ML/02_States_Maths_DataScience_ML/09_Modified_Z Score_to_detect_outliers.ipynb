# Modified Z-Score.

What is modified z-score and how it is used to detect outliers in data science.

We will understand it by an example.

A simple use case of doing study on people height. Here i have 7 different people with their height and the mean/average number is is very simple the sum of all these numbers divided by the number count:
$$Mean = \frac{Sum.of.all.numbers}{numbers.count} $$

So mean/average height of all these people in figure **(09_fig)** will be **6.19.**

There is another concept called median. Median is just a middle number, when you sort all these heights like sorted list and the middle number in that sorted list will be called median **(09_fig1).**
$Median = Middle-Number-in-sorted-list $

Lets look at a different dataset where we do some property price analysis and these are property analysis in **09_fig2** and the analysis im doin is to detect an outlier.

Here in that figure 300k is an outlier, it is like abnormally higher value, see outlier is very subjective, in mant cases you want to consider this as a normal value this is probably not an outlier, so based on case to case you might want to treat these high values as normal values or outlier.

Lets say here you want to treat it as an outlier. You can use mean as your bases i-e 100k and mean you can see in the figure **09_fig3** is impacted by these outliers. If you did not have these three outliers then this mean would have shift somewhere there where most of the values are concentrated but because of these high values the mean is shifted towards the right and standard deviation is nothing but some boundaries around your mean, so one SD for a simple explanation would be some boundaries left and right hand side boundaries around your mean. The usual practise is to use 3 standard dev boundaries to detect outliers. Now here in this case that 300k data point will come under the boundary, so your methadology will not detect this as an outlier and the main reason this is happening is because mean has been affected by this high value, it has been shifted towards right, shown in fig **(09_fig4)**.

What if for the same use case we use median? Median is just a middle number, so median doesnt get impacted that much by these high values and what if we have similar standard deviation and you have a similar kind of a median daviation we call it median absolute deviation and we will look at the mathematical formula, it is similar to standard deviation but it is median instead of mean as it basis point. You can have 2 median absolute deviation, it is knows as **MAD** in short form or 3 or so on, and the usual criteria for MAD is 3.5 and that is the criteria for modified z-score and if we used this kind of methadology for this use-case then it will detect this 300k datapoint as an outlier and the reason about this is that we are using median instead of mean as a base and median is not affected by these high values as shown in **09_fig5**.

The mathematical formula for MAD is that you take every data point subtract it from median of the entire series take the absolute value and take a median of that:
$$MAD = median(|x - median(x)|) $$.

The modified Z-Score is x is your single data point because modified zscore is calculated for every single data point so here x is that single data point minus median of x divided by MAD multiplied by 0.6745 we dont know but it is how the formula is derived:
$$Modified-Z-Score = 0.6745*\frac{X-median(x)}{MAD} $$

Now just lets use modified z-score for the same dataset and lets see what it tells us so for that first look at our formula:
$$MAD = median(|x - median(x)|) $$.

Which one is better z-score or modified z-score?

See if the sample size is smaller and if you have outliers then it will have more bigger impact on average and std and using z-score for that often doesnt work in this situation if we used modified z-score then at least it will detect or tell you or indicate some bigger values and based on that we can make decisions and catch outliers as in previous example.

Now lets write python code.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('09_movie_revenues.csv')
df.head()
```

```python
# if we look at the movies revenues they are such huge numbers so lets first make those into millions or billions.
df.revenue.describe()
```

```python
df['revenue_mln'] = df['revenue'].apply(lambda x: x/1000000)
df.revenue_mln.describe()
```

```python
df.head(2)
```

```python
# lets first apply simple z-score to detect outliers and to compute z-score we need average value and stdev and to get that we need to describe the data set as we did earlier
_, mean, std, *_=df.revenue_mln.describe()
```

```python
mean,std
```

```python
# now we will create a new column in df which will be a z-score of revenue and in order to do that we need to write a function i-e
def get_z_score(value,mean,std):
    return (value-mean)/std
```

```python
df['z_score']=df.revenue_mln.apply(lambda x: get_z_score(x,mean,std))
df.head()
```

```python
df[df.z_score>3] # outlier
```

```python
# lets use modified z-score to detect outliers
```

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df.revenue_mln)
median = np.median(df.revenue_mln)
MAD, median
```

```python
def get_modified_z_score(x,median,MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2787,median,MAD)
```

```python
df['mod_z_score']=df.revenue_mln.apply(lambda x:get_modified_z_score(x,median,MAD))
df.head(3)
```

```python
df[df.mod_z_score>3.5]
```
