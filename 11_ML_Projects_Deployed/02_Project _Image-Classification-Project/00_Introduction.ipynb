# End to End Data science and Machine Learning Project | Image classfication

The Goal of this project to build a website where we can drag and drop an image of a sports person and it will tell us that which sports person that is.

We are taking only taking 5 classes of celebrity or sportpersons dataset.

Images are scrapped from google.

Process:

1. Web scrapping
2. Feature engineering
3. OpenCV for image processing
4. Wavelet transform such as using images to train SVM classifiers or logistic regression classifier

Model Architecure:
1. Building a model
2. Hyper tuning
3. Export
4. Python Flask Server
5. Website | Html, css, javascript
