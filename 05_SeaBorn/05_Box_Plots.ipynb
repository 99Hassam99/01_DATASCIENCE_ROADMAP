# Seaborn BOX PLOTS

The Box Plots are also known as the categorial distribution plots because they help us visualize the different categories of the data very carefully.

A way of displaying the distribution of data based on summary of statistics. Like mean median etc.

It tells the values of outliers. It also help the distribution of data. It also helps in skewness of data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pgn = sns.load_dataset('penguins')
pgn.head()
```

```python
pgn.describe()
```

# Basic BoxPlot

```python
sns.boxplot(pgn.bill_length_mm)
```

Island vs bill_length_mm

```python
sns.boxplot(x=pgn.bill_length_mm, y=pgn.island)
```

# Hue attribute as Sex

```python
sns.boxplot(x=pgn.bill_length_mm, y=pgn.island,hue=pgn.sex)
```

# Vertical instead of Horizontal bar

```python
sns.boxplot(y=pgn.bill_length_mm, x=pgn.island,hue=pgn.sex)
```

# Ordering of Bars

```python
sns.boxplot(y=pgn.bill_length_mm, x=pgn.island,hue=pgn.sex, order= ['Biscoe','Dream','Torgersen'])
```

# Color attribute

```python
sns.boxplot(y=pgn.bill_length_mm, x=pgn.island,hue=pgn.sex, order= ['Biscoe','Dream','Torgersen'],color='y')
```

# Width attribute

Not recommended

```python
sns.boxplot(y=pgn.bill_length_mm, x=pgn.island,hue=pgn.sex, order= ['Biscoe','Dream','Torgersen'], color='g', width=1.5)
```

# Whis property

```python
sns.boxplot(y=pgn.bill_length_mm, x=pgn.island,hue=pgn.sex, order= ['Biscoe','Dream','Torgersen'], whis=0.5)
```

# Showcap

```python
sns.boxplot(y=pgn.bill_length_mm, x=pgn.island,hue=pgn.sex, order= ['Biscoe','Dream','Torgersen'], whis=0.5, showcaps=False)
```
