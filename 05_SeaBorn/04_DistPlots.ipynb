# Displots

Displots in seaborn are made up of 3 subsets.

1.KDE plot

2.Normal Histogram plot

3.Rug Plot

Distplots are also known as distribution plots.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pgn = sns.load_dataset('penguins')
pgn.head()
```

# Drawing a simple Distplot

```python
sns.distplot(pgn.flipper_length_mm, rug = True)
```

The y axis give the probability density function of data.

# Removing the KDE & HIST Plot

```python
sns.distplot(pgn.flipper_length_mm, kde = False)
```

```python
sns.distplot(pgn.flipper_length_mm, hist = False)
```

# Setting number of bins

```python
sns.distplot(pgn.flipper_length_mm, bins=20)
```

# Setting the bin values as an array

```python
sns.distplot(pgn.flipper_length_mm, bins=[160,170,180,190,200,220])
```

It is not recommended

Fitting the data to a particular dist using norm_hist

```python
sns.distplot(pgn.flipper_length_mm, norm_hist=True, kde = False)
```

# Styling each RUGPLOTS

```python
sns.distplot(pgn.flipper_length_mm, rug_kws = {
    'color': 'orange',
    'height':0.20
},rug=True)
```

# Styling Histogram

```python
sns.distplot(pgn.flipper_length_mm,kde= False, hist_kws={
    'color':'red',
    'alpha':0.20,
    'histtype':'step'
})
```

```python
sns.distplot(pgn.flipper_length_mm,kde= False, hist_kws={
    'color':'red',
    'alpha':0.20,
    'histtype':'stepfilled'
})
```

# Styling KDE Plots

```python
sns.distplot(pgn.flipper_length_mm,kde= True, hist= False, kde_kws={
    'color':'blue',
    'lw':4.0,
    'alpha':0.4
})
```
