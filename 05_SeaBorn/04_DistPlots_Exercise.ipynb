```python
# Import the rrequired dependencies to draw seaborn plots
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the mpg dataset and store it in a variable called cars
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
# Draw a distplot for the displacement column of the mpg dataset

sns.distplot(mpg.displacement, rug = True)
```

```python
# Disable the KDE Curve from the above plot using required the required function

sns.distplot(mpg.displacement, rug = True, kde=False)
```

```python
# Enable the rugplot in the above plot using required function

sns.distplot(mpg.displacement,kde=False,rug = True)
```

```python
# Set the number of bins to 20 in above plot

sns.distplot(mpg.displacement, kde=False, rug=True, bins=20)
```

```python
# Convert the plot above to a normal distribution with no of bins as 10 
from scipy.stats import norm
sns.distplot(mpg.displacement,kde=False, rug=True, fit=norm)
```

```python
# Change the transperency of the histogram in the above plot to 0.6 and also the hist type to step
sns.distplot(mpg.displacement, kde=False,rug=True, hist_kws={
    'histtype':'step',
})
```

```python
# Change the color of the above plot to red

sns.distplot(mpg.displacement, kde=False,rug=True, hist_kws={
    'color':'red',
    'histtype':'step',
})
```

```python
# Draw a distplot with the displacement as the column and change the line width of the KDE Plot to 5.
sns.distplot(mpg.displacement, kde=True, kde_kws={
    'lw':5
})
```

```python
# In the above plot change the color of the rugplot to red and the height

sns.distplot(mpg.displacement, rug=True, rug_kws={
    'color':'red',
    'height':0.15

})
```
