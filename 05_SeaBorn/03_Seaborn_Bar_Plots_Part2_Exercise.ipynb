```python
# Import the necessary dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the penguins dataset
penguin = sns.load_dataset('penguins')
```

```python
# Check the first five rows of the dataset using the head function of Pandas
penguin.head()
```

```python
# Draw a bar plot of species vs bill_length_mm

sns.barplot(x = 'species', y='bill_length_mm', data = penguin)
```

```python
# Change the estimator on the y axis to median in the above plot

sns.barplot(x = 'species', y='bill_length_mm',estimator = np.median, data = penguin)
```

```python
# Draw a bar plot of sex vs flipper_length_mm

sns.barplot(x='sex',y='flipper_length_mm',data=penguin)
```

```python
# Change the confidence interval of the above plot to 100

sns.barplot(x='sex',y='flipper_length_mm', errorbar=('ci', 100), data=penguin)
```

```python
# Set the capsize of the above plot to 0.2 with the same confidence interval of 100

sns.barplot(x='sex',y='flipper_length_mm', errorbar=('ci', 100),capsize=0.2, data=penguin)
```

```python
# Set the color of the bins to blue in above plot

sns.barplot(x='sex',y='flipper_length_mm', color='red', data=penguin)
```

```python
# Set the saturation of the blue in the above plot to 0.4

sns.barplot(x='sex',y='flipper_length_mm', color='red',saturation=0.4, data=penguin)
```

```python
# Draw a plot of island vs bodymass_g

sns.barplot(x='island',y='body_mass_g',data=penguin)
```

```python
# Set the palette of the above plot to spring

sns.barplot(x='island',y='body_mass_g',palette='spring',data=penguin)
```

```python
# Save the plot with name as body_mass vs island.png using the appropriate Seaborn functions

sns_plot=sns.barplot(x='island',y='body_mass_g',data=penguin)
```

```python
fig = sns_plot.get_figure()
```

```python
fig.savefig(fname = '03_body_mass vs island.png')
```
