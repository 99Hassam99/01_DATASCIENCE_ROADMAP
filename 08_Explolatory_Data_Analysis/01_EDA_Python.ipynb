# What is EDA?

EDA is nothing but a data exploration technique to understand the various aspects of the data. It includes several techniques in a sequence that we have to follow. Understanding the data includes alot of steps when we are exploring a data, so fews things should be kept in mind while exploring the data like we have to make sure that the data is clean and does not have any redundancies or missing values or even null values in dataset and we have to make sure that we identify the important variables in the dataset and remove all the unnecessary noise in the data that may handle the accuracy of our conclusion when we do model building and we must understand the relationship between the variables through EDA and last but not the least we must be able to derive the conclusions or gather insights about the data or conclusive interpretation in order to move on to more complex processes in the data processing life cycle.

# What are the objectives of EDA?

It is basically used to filter the data from redundancied means the very basic idea is to make sure that data after EDA is clean and by clean means the data has to be free from all the redundancies including null values etc so it can be divided into 2 basic objectives to perform EDA:

1) EDA helps us in identifying the faulty points in data and if you have identify the faulty points in data then you can easily remove them and clean your data.
2) EDA helps us to understand the relationship between the variables which gives us the wider perspective on the data and it actually helps us build on it by utilizing the relationship between the variables.

So these are the main objectives of performing EDA on any DATA.

# Which steps are involved in EDA?

It follows a systematic set of steps to explore the data in the most efficient way possible:

1) Understand the data: The very 1st step is to understand the variables in the dataset so you have to be sure about what kind of data you are working at or what are the variables like numbers,columns and rows etc and how it actually looks like that is the 1st step after loading the data in the program.
2) Cleaning the data: The 2nd step is to clean the data from redundancies, redundancies can be irregularities in the dataset, it can be some variables or some columns that are not necessary for making our conclusions or interpretations so we can just remove them or they are outliers which can cause noise in the data.
3) Analyse the relation between variables: The last but not the least we have to analyse the relationship between variables
