```python
new_df = df.interpolate(method = 'time')
new_df.head()
```

```python
new_df=df.ffill()
new_df
```

```python
new_df.isnull().sum()
```

```python
new_df2=new_df.bfill()
new_df2
```

```python
new_df2.replace("UNKNOWN", np.nan, inplace=True)
```

```python
new_df2.head()
```

```python
new_df3 = new_df2.bfill()
new_df3
```

```python
new_df3.isnull().sum()
```
